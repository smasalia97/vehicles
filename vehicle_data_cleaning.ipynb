```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("vehicles.csv")
df.head()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
print(df.shape)
df.info()
```

```python
print(df['size'].unique())
df['size'].value_counts(dropna=False)
```

```python
df = df.drop(columns = ['county', 'size', 'region_url'])
df.columns
```

```python
df = df.drop_duplicates(subset='id')  # Drop based on unique ID
```

```python
df.head()
```

```python
null_percent = df.isnull().mean().sort_values(ascending=False)
null_percent
```

```python
df.manufacturer = df.manufacturer.str.title()
df.model = df.model.str.title()

df['vehicle'] = np.where(
    df['manufacturer'].notna() & df['model'].notna(),
    df['manufacturer'].astype(str) + ' - ' + df['model'].astype(str),
    None
)
```

```python
df['year'] = df['year'].fillna(df['year'].median())
df['odometer'] = df['odometer'].fillna(df['odometer'].median())
```

```python
df.posting_date
```

```python
df['posting_date'] = pd.to_datetime(df['posting_date'], utc=True).dt.tz_convert(None)
```

```python
df.describe()
```

```python
q_low = df['price'].quantile(0.05)
q_high = df['price'].quantile(0.95)
df = df[(df['price'] >= q_low) & (df['price'] <= q_high)]
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df = df[['id', 'url', 'region', 'price', 'year', 'vehicle', 'manufacturer',
       'model', 'condition', 'cylinders', 'fuel', 'odometer', 'title_status',
       'transmission', 'VIN', 'drive', 'type', 'paint_color', 'image_url',
       'description', 'state', 'lat', 'long', 'posting_date']]
```

```python
df.cylinders.value_counts(dropna=False)
```

```python
def extract_cylinders(val):
    if pd.isna(val) or val == 'other':
        return None
    try:
        return int(str(val).split()[0])
    except:
        return None

df['cylinders'] = df['cylinders'].apply(extract_cylinders)
df['cylinders'].value_counts(dropna=False)
```

```python
df.head()
```

```python
df.state = df.state.str.upper()
```

```python
df.tail()
```

```python
df.info()
```

```python
df['price_per_mile'] = np.where(
    df['odometer'].notna() & (df['odometer'] != 0),
    df['price'] / df['odometer'],
    np.nan  # or 0, depending on your logic
)
df.head()
```

```python
df.reset_index(drop=True, inplace=True)
df['id'] = df.index + 1  # Start from 1 instead of 0
df['id'] = df['id'].astype(int)  # Ensure it's an integer type
df.head()
```

```python
df.to_csv('cleaned_vehicles.csv', index=False)
```
